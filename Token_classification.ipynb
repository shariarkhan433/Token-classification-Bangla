# **Dataset processing**

```python
from datasets import load_dataset
import re
def remove_punctuation(sentence):
    sentence[data_column] = re.sub(r'[^\u0980-\u09FF\s]', '', sentence[data_column])
    return sentence
dataset = load_dataset('shahidul034/text_generation_model_data')
# dataset = load_dataset('csebuetnlp/xnli_bn')
data_column = 'content'
dataset['train']=dataset['train'].select([x for x in range(0,5000)])
dataset['train']=dataset['train'].filter(lambda x : len(x[data_column])>10)
dataset['train']=dataset['train'].map(remove_punctuation)
```

```python
from datasets import load_from_disk
ner_pos = load_from_disk('/content/drive/MyDrive/ner_pos')
```

```python
ner_pos_updated = ner_pos.map(lambda example: {'tag_num': example['tag_num'] - 1})
```

```python
ner_pos_updated['ner']['tag_num']
```

```python
def assign_tag_pos(tag):
  for x1,x2 in zip(ner_pos_updated['pos']['pos_list'],ner_pos_updated['pos']['tag_num']):
    if tag==x1:
      return x2
def assign_tag_ner(tag):
  for x1,x2 in zip(ner_pos_updated['ner']['ner_list'],ner_pos_updated['ner']['tag_num']):
    if tag==x1:
      return x2
```

```python
from bnlp import BengaliNER
from bnlp import BengaliPOS
bn_pos = BengaliPOS()
bn_ner = BengaliNER()
```

```python
from normalizer import normalize
def fun(ex):
  ex['pos_tags']=[assign_tag_pos(x[1]) for x in (bn_pos.tag(normalize(ex[data_column])))]
  ex['pos']=[x[1] for x in (bn_pos.tag(normalize(ex[data_column])))]
  ex['ner_tags']=[assign_tag_ner(x[1]) for x in (bn_ner.tag(normalize(ex[data_column])))]
  ex['ner']=[x[1] for x in (bn_ner.tag(normalize(ex[data_column])))]
  ex['tokens'] = [x[0] for x in (bn_ner.tag(normalize(ex[data_column])))]
  return ex
tagged_raw_data=dataset['train'].map(fun)
```

```python
tagged_raw_data[0]
```

```python
from transformers import AutoTokenizer

# model_checkpoint = "bert-base-cased"
# model_checkpoint ='csebuetnlp/banglabert'
model_checkpoint = "Shariar433/bert-finetuned-ner-bangla"
tokenizer = AutoTokenizer.from_pretrained(model_checkpoint)
```

```python


def tokenize_and_align_labels(examples): #only for ner
    tokenized_inputs = tokenizer(examples["tokens"], padding=True, is_split_into_words=True, max_length=128, truncation=True)

    labels = []
    for i, label in enumerate(examples[f"ner_tags"]):
        word_ids = tokenized_inputs.word_ids(batch_index=i)
        previous_word_idx = None
        label_ids = []
        for word_idx in word_ids:
            if word_idx is None:
                label_ids.append(-100)
            elif word_idx != previous_word_idx:
                label_ids.append(label[word_idx])
            else:
                label_ids.append(-100)
            previous_word_idx = word_idx

        labels.append(label_ids)
    tokenized_inputs["labels"] = labels
    return tokenized_inputs
```

```python
tokenized_dataset = tagged_raw_data.map(tokenize_and_align_labels, batched=True)
```

```python
from transformers import DataCollatorForTokenClassification

data_collator = DataCollatorForTokenClassification(tokenizer=tokenizer)
```

```python
label_list = ner_pos_updated['ner']['ner_list']
label_list
```

```python
id2label = {str(i): label for label, i in zip(ner_pos_updated['ner']['ner_list'],ner_pos_updated['ner']['tag_num'])}
label2id = {label : i for label, i in zip(ner_pos_updated['ner']['ner_list'],ner_pos_updated['ner']['tag_num'])}
```

```python
id2label
# label2id
```

```python
from transformers import AutoModelForTokenClassification

model = AutoModelForTokenClassification.from_pretrained(
    model_checkpoint,
    id2label=id2label,
    label2id=label2id,
    num_labels=16
)
```

```python
model.config.num_labels
```

```python
dataset = tokenized_dataset.train_test_split(test_size=1000, shuffle = True)
```

```python
dataset['train'].features
```

```python
from transformers import TrainingArguments
from transformers import Trainer

training_args = TrainingArguments(
    output_dir="bert-finetuned-ner-bangla",
    learning_rate=2e-5,
    per_device_train_batch_size=64,
    per_device_eval_batch_size=64,
    num_train_epochs=2,
    weight_decay=0.01,
    evaluation_strategy="steps",
    eval_steps=500,
    save_steps=1000,
    save_strategy="steps",
    load_best_model_at_end=True,
    # push_to_hub=True,
    remove_unused_columns=True,
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=dataset["train"],
    eval_dataset=dataset["test"],
    tokenizer=tokenizer,
    data_collator=data_collator,
    # compute_metrics=compute_metrics,
)

trainer.train()
```

```python
trainer.save_model('bangla-ner')
```
